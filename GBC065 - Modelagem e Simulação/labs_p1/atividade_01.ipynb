#### Exercício 01
Uma caixa contém duas bolas, que podem ser ambas vermelhas, ambas azuis, ou uma de cada cor. A cada passo, uma bola é sorteada e substituída por uma nova bola, que tem 80% de chance de ser da mesma cor. Se inicialmente ambas são vermelhas, qual a probabilidade de que a quinta bola retirada seja vermelha?

```python
# quero P(A5 | X0 = 2) então há duas vermelhas inicialmente
def simulacao_ex01(nro_simulacoes):
    a5 = 0
    for _ in range(nro_simulacoes):
        bolas_v = 2

        for i in range(5):
            bolas = ["vermelha"] * bolas_v + ["azul"] * (2 - bolas_v)

            sorteada = np.random.choice(bolas)

            if i == 4: # se é a quinta, retorna que a bola foi vermelha
                if sorteada == "vermelha":
                    a5 += 1
                break
        
        # removendo bola sorteada e adicionando uma nova
            if sorteada == "vermelha": 
                bolas_v -= 1
                if np.random.random() <= 0.8:
                    bolas_v += 1
            else: # aqui é se for azul
                if np.random.random() > 0.8:
                    bolas_v += 1
    
            bolas_v = max(0, min(2, bolas_v))
            
    return a5/nro_simulacoes

result0 = simulacao_ex01(simulacoes1)
print(f"Com {simulacoes1} simulações há P(A5 | X0 = 2) = {result0:.6f}")

result01 = simulacao_ex01(simulacoes2)
print(f"Com {simulacoes2} simulações há P(A5 | X0 = 2) = {result01:.6f}")
```

#### Exercício 02
Numa estrada há carros e caminhões. Três de cada quatro caminhões na estrada são seguidos por um carro, enquanto que apenas um de cada cinco carros é seguido por um caminhão. Qual a porcentagem de caminhões na estrada?

```python
def simulacao_ex02(nro_simulacoes):
    cont_caminhoes = 0

    # caminhoes são os valores 1 e carros são os valores 0
    # começando com carro
    atual = 1

    # esses números eu obti pelo enunciado: 3/4 = 0.75 e 1/5 = 0.2
    prob_caminhao_carro = 0.75
    prob_carro_caminhao = 0.2

    for _ in range(nro_simulacoes):
        if atual == 1: 
            cont_caminhoes += 1

            if np.random.random() < prob_caminhao_carro: # decidindo próximo veículo dado que atual é caminhão
                atual = 0
            else:
                atual = 1
        else:
            if np.random.random() < prob_carro_caminhao: # decidindo próximo veículo dado que atual é carro
                atual = 1
            else:
                atual = 0
    
    return (cont_caminhoes / nro_simulacoes) * 100 # porcentagem de caminhões na estrada

result1 = simulacao_ex02(simulacoes1)
print(f"Com {simulacoes1} simulações há uma porcentagem de {result1:.6f}% de caminhões.")
result2 = simulacao_ex02(simulacoes2)
print(f"Com {simulacoes2} simulações há uma porcentagem de {result2:.6f}% de caminhões.")
```

#### Exercício 03
Sabe-se que cada item produzido por uma certa máquina será defeituoso com uma probabilidade 0,001, independentemente de qualquer outro item. Qual a probabilidade de que em uma amostra contendo três itens, pelo menos um esteja com defeito?

```python
def simulacao_ex03(nro_simulacoes):
    prob_defeito = 0.001 
    nro_defeitos = 0 # quantos defeitos foram encontrados no total

    for _ in range(nro_simulacoes):
        defeituosa = False # começa com nenhuma peça defeituosa encontrada ainda

        for _ in range(3):
            if np.random.random() < prob_defeito:
                defeituosa = True # se encontrou, para a busca na amostra
                break

        if defeituosa:
            nro_defeitos += 1
    
    return nro_defeitos/nro_simulacoes

result4 = simulacao_ex03(simulacoes1)
print(f"Com {simulacoes1} simulações há uma probabilidade de {result4:.6f} de pelo menos um item defeituoso.")
result5 = simulacao_ex03(simulacoes2)
print(f"Com {simulacoes2} simulações há uma probabilidade de {result5:.6f} de pelo menos um item defeituoso.")
```
