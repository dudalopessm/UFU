Distribuição de resultados de um dado D20

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500000  # número de lançamentos

dado = np.random.randint(1, 21, N)  # N valores entre 1 e 20

media = dado.mean()  # valor médio dos valores aleatórios -> valor esperado
print(media)

plt.hist(dado, bins=20) # histograma de distribuição dos números aleatórios
```

Distribuição de resultados de um lançamento de um dado D20 com vantagem

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500000  # número de lançamentos

dado1 = np.random.randint(1, 21, N)  # N valores entre 1 e 20
dado2 = np.random.randint(1, 21, N)  # N valores entre 1 e 20

dadofinal = np.maximum(dado1, dado2) # máximo entre os dois dados, pois o lançamento com vantagem é assim

media = dadofinal.mean()  # valor médio dos valores aleatórios -> valor esperado
print(media)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(dadofinal, bins=20)  # histograma de distribuição dos números aleatórios

# plotando a frequência no histograma
dado_string = [str(dado) for dado in dadofinal]
plt.subplot(1, 2, 2)
plt.hist(dado_string, bins=20)
```

**PROBLEMA DE PROBABILIDADE**

Suponha que temos uma caixa com 5 bolas pretas e 7 bolas brancas.
Compramos duas sem colocar de volta. Qual é a distribuição de
probabilidade da variável aleatória X = número de bolas pretas
retiradas? E qual é o número esperado de bolas pretas retiradas?

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1_000_000

X = [] # variável aleatória X

for _ in range(N):
  pretas = 5
  brancas = 7
  for  _ in range(2):
    total = pretas + brancas
    bola = np.random.randint(1, total+1)
    if (bola <= pretas):
      pretas -= 1
    else:
      brancas -= 1
  X.append(5 - pretas)

plt.hist(X, bins=20)
print(np.mean(X))
```
