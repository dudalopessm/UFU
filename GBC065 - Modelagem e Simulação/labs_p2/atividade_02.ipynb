01. Helena está na janela de casa observando uma tempestade de raios
no horizonte. Um raio cai em média a cada 45 segundos. Se Helena
não vê nenhum raio no primeiro minuto, ela cansa de esperar e sai da
janela. Senão, ela observa por mais dois minutos.

(a) Qual o número esperado de raios que Helena observa?

(b) Qual o tempo esperado que ela fica na janela?

```python
import numpy as np

# Lambda = 1/média_segundos = 1/45
# Quando t = 0, ela observa. Se não cair um raio nos próximos 60s, ela sai. Se cair, ela observa por mais dois minutos, ou seja, fica 180s na janela.

experimentos = 100000

lamb = 1/45

t_inicial = 0
t_final = 180

raios_vistos_total = 0
tempo_total = 0

esperados = lamb*(t_final - t_inicial)

for i in range(experimentos):
  # Total de raios em 3 minutos
  eventos = np.random.poisson(esperados)

  tempo_eventos = np.sort(np.random.uniform(t_inicial, t_final, size=eventos))

  um_min = 0
  total = 0

  # Contagem eventos
  for j in tempo_eventos:
    if j <= 60:
      um_min += 1
    if j <= 180:
      total += 1

  # Se não viu em 60s, sai da janela
  if um_min == 0:
    observacao = 60
    raios_vistos = 0
  else:
    # Se viu, ela ficou 180s na janela
    observacao = 180
    raios_vistos = total

  raios_vistos_total += raios_vistos
  tempo_total += observacao

print(f"Número esperado de raios observados: {(raios_vistos_total/experimentos):.3f} raios.")
print(f"Tempo esperado que Helena fica na janela (s): {(tempo_total/experimentos):.3f} segundos.")
```

2. Cem itens são simultaneamente colocados num teste de vida. Suponha que os tempos de vida de cada um dos itens são variáveis independentes e que seguem uma distribuição exponencial com média 200 horas. O teste termina quando há um total de 5 falhas. Qual o tempo esperado da duração do teste?

```python
import numpy as np

# Lambda = 1/média_horas = 1/200 = 0.005 falhas por hora
# Lambda_sistema = n_itens * lambda = 100*0.005 = 0.5
# São 5 falhas para o teste terminar, então o tempo total do teste é: T = T1 + T2 + T3 + T4 + T5 em que cada T = exp^taxa_atual

experimentos = 100000
media_horas = 200
lamb_item = 1 / media_horas

num_itens = 100
falhas_necessarias = 5

tempos_totais = []

for i in range(experimentos):
    itens_vivos = num_itens
    tempo_total = 0

    # Simula até ocorrerem 5 falhas
    for f in range(falhas_necessarias):
        taxa_sistema = itens_vivos * lamb_item

        # Tempo até próxima falha
        tempo_falha = np.random.exponential(1 / taxa_sistema)

        tempo_total += tempo_falha

        # Item falhou
        itens_vivos -= 1

    tempos_totais.append(tempo_total)

print(f"Tempo médio de duração do teste: {np.mean(tempos_totais):.3f} horas.")
```
